## Getting Model from Unsloth Model Space Or Huggingface Transformers

```python
# Importing "FastLanguageModel" from Unsloth package. Used to load pretrained large language models (LLM)
from unsloth import FastLanguageModel

max_seq_length = 2048
dtype = None
load_in_4bit = True


model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/DeepSeek-R1-Distill-Llama-8B",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
    # token = hf_token,     # Uncomment and pass Huggingface token when prompted if the Model is secured and private
)
```

```python
prompt_style = """Below is an instruction that describes a task, paired with an input that provides further context.
Write a response that appropriately completes the request.
Before answering, think carefully about the question and create a step-by-step chain of thoughts to ensure a logical and accurate response.

### Instruction:
You are a Snowflake expert with entire advanced knowledge in Cloud Data Warehousing.
Please answer the following Snowflake question.

### Question:
{}

### Response:
<think>{}"""
```

## Querying model with a simple question

```python
question = "Generate a Snowflake SQL query to list all the users that have access to the database namedÂ marketing."


FastLanguageModel.for_inference(model)
inputs = tokenizer([prompt_style.format(question, "")], return_tensors="pt").to("cuda")

outputs = model.generate(
    input_ids=inputs.input_ids,
    attention_mask=inputs.attention_mask,
    max_new_tokens=1200,
    use_cache=True,
)
response = tokenizer.batch_decode(outputs)
print(response[0].split("### Response:")[1])
```

# Loading Snowflake Instruction dataset from Local space

```python
from datasets import load_from_disk
dataset = load_from_disk("/content/sflk_instruction.hf")
dataset
```

## Making sample dataset from Original dataset to query model with questions available in this sample dataset

```python
sample_dataset = dataset.train_test_split(test_size=0.1)['test'] # Performs data split into train and test samples
sample_dataset
```

### Prettifying the sample dataset into Tabular format

```python
import pandas as pd    # Data handler library to perform data related operations
sample_dataframe = pd.DataFrame(sample_dataset)
sample_dataframe.head(5)
```

## Generating Answers for Question available in our sample dataframe.
#### **Our loaded foundation model is still untrained. To capture the responses, we inferencing the model querying before finetuing or training it with out datset**

```python
# Defining a function to generate answers in a batch
def generate_batch_answers(question):
    # question = "Generate a Snowflake SQL query to list all the users that have access to the database namedÂ marketing."

    FastLanguageModel.for_inference(model)
    inputs = tokenizer([prompt_style.format(question, "")], return_tensors="pt").to("cuda")

    outputs = model.generate(
        input_ids=inputs.input_ids,
        attention_mask=inputs.attention_mask,
        max_new_tokens=1200,
        use_cache=True,
    )
    response = tokenizer.batch_decode(outputs)
    # print(response[0].split("### Response:")[1])
    return response[0].split("### Response:")[1]
```

```python
sample_dataframe['before_tuning'] = sample_dataframe['Instruction'].apply(generate_batch_answers)

sample_dataframe.head(3)
```

# **Starting our Model Training**

---



## Step 1. We re-initiating our loaded model with PEFT(Parameter Efficient Fine Tuning) config parameters.

Parameters used

*   r
*   target_modules
*   lora_alpha
*   lora_dropout
*   bias
*   gradient_checkpoint
*   random_state
*   rs_lora
*   loftq_confg











```python
model = FastLanguageModel.get_peft_model(
    model,
    r=16,
    target_modules=[
        "q_proj",
        "k_proj",
        "v_proj",
        "o_proj",
        "gate_proj",
        "up_proj",
        "down_proj",
    ],
    lora_alpha=16,
    lora_dropout=0,
    bias="none",
    use_gradient_checkpointing="unsloth",  # True or "unsloth" for very long context
    random_state=3407,
    use_rslora=False,
    loftq_config=None,
)
```

# 2. Formatting our dataset into Prompt template structure

---


### Appending our prompted data format as a new field **'text'** into our original dataset

Technical terms used
*   EOS_TOKEN
*   tokenizer.eos_token



```python
EOS_TOKEN = tokenizer.eos_token  # Must add EOS_TOKEN


def formatting_prompts_func(examples):
    inputs = examples["Instruction"]
    cots = examples["Input"]
    outputs = examples["Output"]
    texts = []
    for input, cot, output in zip(inputs, cots, outputs):
        text = prompt_style.format(input, cot, output) + EOS_TOKEN
        texts.append(text)
    return {
        "text": texts,
    }

dataset = dataset.map(formatting_prompts_func, batched = True,)
dataset["text"][0]
```

## 3. Declaring Model Training with Supervised FineTuning concept (SFT)
---
Parametes used
*   model
*   tokenizer
*   train_dataset
*   dataset_text_field
*   max_seq_length
*   dataset_num_proc
*   TrainingArguments
    *   per_device_train_batch_size
    *   gradient_accumulation_steps
    *   warmup_steps
    *   max_steps
    *   learning_rate
    *   fp16
    *   bf16
    *   logging_steps
    *   optim
    *   weight_decay
    *   lr_scheduler_type
    *   seed
    *   output_dir


```python
from trl import SFTTrainer
from transformers import TrainingArguments
from unsloth import is_bfloat16_supported

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=dataset,
    dataset_text_field="text",
    max_seq_length=max_seq_length,
    dataset_num_proc=2,
    args=TrainingArguments(
        per_device_train_batch_size=2,
        gradient_accumulation_steps=4,
        # Use num_train_epochs = 1, warmup_ratio for full training runs!
        warmup_steps=5,
        max_steps=60,
        learning_rate=2e-4,
        fp16=not is_bfloat16_supported(),
        bf16=is_bfloat16_supported(),
        logging_steps=10,
        optim="adamw_8bit",
        weight_decay=0.01,
        lr_scheduler_type="linear",
        seed=3407,
        output_dir="outputs",
        report_to="none",
    ),
)
```

## 4. Calling Trainer class to initiate model training (actual finetuning happens here)

---



```python
trainer_stats = trainer.train()
```
